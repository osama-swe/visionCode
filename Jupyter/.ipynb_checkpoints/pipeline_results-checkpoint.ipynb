```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import scipy.misc
import glob
import imageio
import os
```

```python
def color_thresh(img, rgb_thresh=(160, 160, 160)):
    color_select = np.zeros_like(img[:,:,0])
    above_thresh = (img[:,:,0] > rgb_thresh[0]) \
                & (img[:,:,1] > rgb_thresh[1]) \
                & (img[:,:,2] > rgb_thresh[2])
    color_select[above_thresh] = 1
    return color_select

def detect_rocks(img, rgb_thresh=(110, 110, 60)):
    color_select = np.zeros_like(img[:,:,0])
    above_thresh = (img[:,:,0] > rgb_thresh[0]) \
                & (img[:,:,1] > rgb_thresh[1]) \
                & (img[:,:,2] < rgb_thresh[2])
    color_select[above_thresh] = 1
    return color_select

def rover_coords(binary_img):
    ypos, xpos = binary_img.nonzero()
    x_pixel = -(ypos - binary_img.shape[0]).astype(np.float64)
    y_pixel = -(xpos - binary_img.shape[1]/2 ).astype(np.float64)
    return x_pixel, y_pixel


def to_polar_coords(x_pixel, y_pixel):
    dist = np.sqrt(x_pixel**2 + y_pixel**2)
    angles = np.arctan2(y_pixel, x_pixel)
    return dist, angles

def rotate_pix(xpix, ypix, yaw):
    yaw_rad = yaw * np.pi / 180
    xpix_rotated = (xpix * np.cos(yaw_rad)) - (ypix * np.sin(yaw_rad))                    
    ypix_rotated = (xpix * np.sin(yaw_rad)) + (ypix * np.cos(yaw_rad))
    return xpix_rotated, ypix_rotated

def translate_pix(xpix_rot, ypix_rot, xpos, ypos, scale): 
    xpix_translated = (xpix_rot / scale) + xpos
    ypix_translated = (ypix_rot / scale) + ypos
    return xpix_translated, ypix_translated


def pix_to_world(xpix, ypix, xpos, ypos, yaw, world_size, scale):
    xpix_rot, ypix_rot = rotate_pix(xpix, ypix, yaw)
    xpix_tran, ypix_tran = translate_pix(xpix_rot, ypix_rot, xpos, ypos, scale)
    x_pix_world = np.clip(np.int_(xpix_tran), 0, world_size - 1)
    y_pix_world = np.clip(np.int_(ypix_tran), 0, world_size - 1)
    return x_pix_world, y_pix_world

def perspect_transform(img, src, dst):           
    M = cv2.getPerspectiveTransform(src, dst)
    warped = cv2.warpPerspective(img, M, (img.shape[1], img.shape[0]))# keep same size as input image
    mask = cv2.warpPerspective(np.ones_like(img[:,:,0]), M, (img.shape[1], img.shape[0]))
    return warped, mask
```

```python
path = '/home/kali/Documents/Vision/Vision_Project_Startup_Code/test_dataset/IMG/*'
img_list = glob.glob(path)
idx = np.random.randint(0, len(img_list) - 1)
image = mpimg.imread(img_list[idx])



dst = 5
bottom_offset = 6
source = np.float32([[14, 140],
                 [300, 140],
                 [200, 95],
                 [120, 95]])
destination = np.float32([[image.shape[1] / 2 - dst, image.shape[0] - bottom_offset],
                        [image.shape[1] / 2 + dst, image.shape[0] - bottom_offset],
                        [image.shape[1] / 2 + dst, image.shape[0] - 2*dst - bottom_offset],
                        [image.shape[1] / 2 - dst, image.shape[0] - 2*dst - bottom_offset]])

warped, mask = perspect_transform(image, source, destination)


threshed = color_thresh(warped, rgb_thresh=(160, 160, 160))
tmp_threshed = color_thresh(warped, rgb_thresh=(120, 120, 120))
obstacles_thresh = (np.ones_like(threshed)-tmp_threshed)*mask
rocks_thresh = detect_rocks(warped)

vision_image = np.zeros((160, 320, 3))
vision_image[:, :, 0] = obstacles_thresh
vision_image[:, :, 1] = rocks_thresh
vision_image[:, :, 2] = threshed

xpix, ypix = rover_coords(threshed)
xobstacles, yobstacles = rover_coords(obstacles_thresh)
xrocks, yrocks = rover_coords(rocks_thresh)

dists, angles = to_polar_coords(xpix, ypix)
mean_dir = np.mean(angles)
```

```python
plt.figure(figsize=(17,11))

plt.subplot(331)
plt.imshow(image)
plt.title('Rover image (step 1)')

plt.subplot(332)
plt.imshow(threshed, cmap='gray')
plt.title('Threshed image for terrain detection (step 3)')

plt.subplot(333)
plt.imshow(obstacles_thresh, cmap='gray')
plt.title('Threshed image for obstacles detection (step 3)')

plt.subplot(334)
plt.imshow(warped)
plt.title('Bird Eye View (step 2)')

plt.subplot(335)
plt.plot(xpix, ypix, '.', color='blue')
plt.ylim(-160, 160)
plt.xlim(0, 160)
arrow_length = 100
x_arrow = arrow_length * np.cos(mean_dir)
y_arrow = arrow_length * np.sin(mean_dir)
plt.arrow(0, 0, x_arrow, y_arrow, color='orange', zorder=2, head_width=10, width=2)
plt.title('Approximate direction where the Rover should move (step 4)')

plt.subplot(336)
plt.plot(xobstacles, yobstacles, '.', color='red')
plt.ylim(-160, 160)
plt.xlim(0, 160)
plt.title('Rover-centric obstacle pixels (step 4)')

plt.subplot(337)
plt.imshow(vision_image)
plt.title('Rover vision image (step 5)')
```
